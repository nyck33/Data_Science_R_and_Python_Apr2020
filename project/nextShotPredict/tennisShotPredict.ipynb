```python
# import dataset
import pandas as pd

game1clip1 = pd.read_csv(
    "/home/nobu/Desktop/stat3990/project/tracknet/Dataset/game1/Clip1/Label.csv")

print(game1clip1.head(10))

g1_c1 = game1clip1.to_numpy()
g1_c1.dtype

g1_c1.shape

y = g1_c1[:, 3]

print(y[:5])
print(np.max(y),
np.min(y),
np.mean(y),
np.median(y))
```

```python
# find peaks

from scipy.signal import find_peaks
#peaks are frame nums of peaks
peaks, props = find_peaks(y, height = 380, threshold=1, distance=25, prominence=50)
troughs, tprops = find_peaks(-y, height=-380, threshold=0,distance=25, prominence=50)

#props gives the actual y coordinates of peaks
print("peaks {}\n heights {}".format(peaks, props['peak_heights']))
print("troughs {}\n heights {}".format(troughs, -tprops['peak_heights']))
plt.plot(y) 
plt.xlabel=('frame')
plt.ylabel=('y_pos')
plt.plot(0,y[0], "x")
plt.plot(len(y)-1,y[-1], "x")

plt.plot(peaks, y[peaks], "x")
plt.plot(troughs, y[troughs], "o")

#plt.plot(380, "--", color='gray')
plt.show()

all_shots = np.sort(np.concatenate((peaks,troughs)))
min_frames = np.min(np.diff(all_shots))
mean_frames = np.mean(np.diff(all_shots))
max_frames = np.max(np.diff(all_shots))
print(all_shots, mean_frames)
```
